```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.array([[1,1],[2,-2],[-1,-1.5],[-2,-1],[-2,1],[1.5,-0.5]])
#Adding [2,0] with d=-1to X makes X not linearly separable
d=[1,-1,-1,-1,1,1]
w=[0,0]
b=0
```

```python
for i in range(100):
    for i in range(len(X)):
        y=np.dot(w,X[i])+b
        if y>=0:
            fy=1
        elif y<0:
            fy=-1
        if d[i]!=fy:
            w=np.add(w,(((d[i]-fy)/2)*X[i]))
            b=b+((d[i]-fy)/2)
            x = np.linspace(-5,5,100)
            y = ((-w[0]/w[1])*x)+((-b)/w[1])
            plt.figure()

            plt.plot(x, y, '-r')
            for i in range(len(X)):
                if d[i]==1:
                    plt.scatter(X[i][0],X[i][1],marker='^',color='#ff7f0e')
                else:
                    plt.scatter(X[i][0],X[i][1],marker='^',color='#17becf')
            plt.show()

print('Updated Weight Vector is',w)
print('Updated Bias is',b)
fy=[]
np.asarray(fy)
print('____________')

for i in range(len(X)):
    y=np.dot(w,X[i])+b
    if y>=0:
        fy=np.append(fy,[1])
    elif y<0:
        fy=np.append(fy,[-1])

flag=0
for i in range(len(X)):
    if d[i]!=fy[i]:
        flag=1
if flag==1:
    print('Convergence not achieved')
else:
    print('Convergence achieved')
```
